# Customer churn data Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

# laod all data

```python
df = pd.read_csv("Customer Churn.csv")
```

```python
df.head()
```

```python
df.info()
```

# Replacing blank with 0 as tenure is 0 and no total charges are recorded 

```python
df["TotalCharges"]= df["TotalCharges"].replace(" ","0")
df["TotalCharges"]= df["TotalCharges"].astype("float")
```

```python
df.isnull().sum().sum()
```

```python
df.describe()
```

# check the dublicate value in data

```python
df["customerID"].duplicated().sum()
```

```python
def conv(value):
    if value ==1:
        return "yes"
    else:
        return "no"
df["SeniorCitizen"]= df["SeniorCitizen"].apply(conv)
```

```python
df.head(5)
```

```python
ax=sns.countplot(x='Churn', data =df,width=0.5, color='magenta')
ax.bar_label(ax.containers[0])
plt.title("Count of Customer by Churn")
plt.show()
```

```python
plt.figure(figsize=( 3,4))
gb = df.groupby('Churn').agg({'Churn':"count"})
plt.pie(gb['Churn'],labels=gb.index,autopct="%1.2f%%")
plt.title("Percentage Of Churned Customeres", fontsize=10)
plt.show()
```

```python
plt.figure(figsize =(3,4))
ax=sns.countplot(x='gender',data= df,hue="Churn")
plt.bar_label(ax.containers[0])
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize =(3,4))
ax=sns.countplot(x="SeniorCitizen", data =df)
plt.bar_label(ax.containers[0])
plt.title("Count of Customer by Senior Citizen")
plt.show()
```

```python
plt.figure(figsize = (3,4))
ax=sns.countplot(x="SeniorCitizen" ,data= df,hue="Churn")
plt.title("SeniorCitizen by Chrn" )
plt.bar_label(ax.containers[0])
plt.bar_label(ax.containers[1])
plt.show()
```

```python
# Assuming df is your DataFrame
# Create a cross-tab to get counts for 'SeniorCitizen' and 'Churn'
ct = pd.crosstab(df['SeniorCitizen'], df['Churn'], normalize='index') * 100  # Normalize to get percentage

# Plot a stacked bar chart
ct.plot(kind='bar', stacked=True, figsize=(3, 4), color=['skyblue', 'orange'])

# Add percentage labels on the stacked bar
for i, row in enumerate(ct.iterrows()):
    for j, value in enumerate(row[1]):
        plt.text(j, value/2 + ct.iloc[i,:j].sum(), f'{value:.1f}%', ha='center', va='center', color='black')

# Set the title and labels
plt.title('Churn by SeniorCitizen ')
plt.ylabel('Percentage (%)')
plt.xlabel('SeniorCitizen')
plt.legend(title='Churn', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

```python
# List of column names to plot
columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

n_cols=3
n_rows=(len(columns)+n_cols-1)// n_cols

fig, axes = plt.subplots(n_rows,n_cols, figsize=(15, n_rows*4))  # 3x3 grid


# Flatten the axes array for easier iteration
axes = axes.flatten()

# Loop through each column and create a countplot in a subplot
for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i],hue="Churn")
    axes[i].set_title(f'Count plot of {col}')
    axes[i].set_xlabel('')

for j in range(i+1,len(axes)):
    fig.delases(axes[j])
# Adjust layout
plt.tight_layout()
plt.show()
```
